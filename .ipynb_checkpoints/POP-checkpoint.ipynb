## Use ```glob``` to grab the names of all the gfs files

````python
grb_path = "blend.t06z.master.f001.co"
grib_list = sorted(glob.glob(f"{grb_path}*.grib2"))
list(grib_list)
````

## Grabbing the first grib file from our list will allow us to explore the data
* Each entry of the file is a different variable (specified at different heights for most)

````python
# grab first file in list (conus)
grbs = pygrib.open(grib_list[0])
grb = grbs.read()
````

````python
list(grb)
````

````python
# play with cape values - second in the list of variables

precip_grib = grb[18]
print(precip_grib)
precip,lat,lon = precip_grib.data() # -> gets all the plot data in one call
#lat,lon = cape_grib.latlons() #-> just different way of accessing lat/lon; same outcome
precip.shape,lat.shape,lon.shape

print(numpy.amax(precip))
precip_grib.parameterName #shows name 
````

````python
fig = plt.figure(figsize=(20, 20))



ax = plt.axes(projection=ccrs.PlateCarree())

#https://www.spc.noaa.gov/exper/sref/srefplumes/
#https://hwp-viz.gsd.esrl.noaa.gov/wave1d/
#https://matplotlib.org/stable/tutorials/colors/colormaps.html 
#https://sites.google.com/ucar.edu/operational-models-encyclo/probabilistic-models/gefs/cloud-effects-on-radiation
#https://www.weather.gov/mdl/nbm_grib2_4_v32#is4 
#What probability is

##Wtf is plate carree
plt.contourf(lon, lat, precip, 60,
            transform=ccrs.PlateCarree())


#21 rain
#22 not working
#23 snow
#24 Ice pellets/freezing rain
#25 ?
#26 ?
#27 thunderstorm

ax.text(-105.2705, 40.015, 'BOULDER ', transform=ccrs.Geodetic()) #Plots Boulder City as a point
ax.plot(-105.3, 40.1, 'bo', markersize=7, transform=ccrs.Geodetic())
ax.coastlines() #Coastlines
#ax.add_feature(cfeatures.BORDERS, linestyle = ':')
##ax.add_feature(borders) country borders
ax.set_extent([-109, -102, 37, 41], crs=ccrs.PlateCarree())
ax.set_aspect(1)
plt.title('Probability of like rain')

plt.show()
````
